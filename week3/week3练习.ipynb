```python
df = pd.read_csv ("20春_pandas_week02_hurun_unicorn.tsv", encoding = "utf8", sep="\t")
df.info()
```

```python
#先行再国 先国再行 先国再城 先行再城

先国再行 = df.groupby ( by = ['国家','行业'] ) \
             .agg ({ "企业名称" : "count", \
                     "估值（亿人民币）":["sum","mean"], \
                     "成立年份":["max","min"],               }) \
             .sort_values ( by = [("估值（亿人民币）","sum")], ascending = False) \
             .rename ( columns = {"sum":"总和", "mean":"均值", "count":"数量", "max":"最新", "min":"最早"} )
先行再国 = df.groupby ( by = ['行业', '国家'] ) \
             .agg ({ "企业名称" : "count", \
                     "估值（亿人民币）":["sum","mean"], \
                     "成立年份":["max","min"],               }) \
             .sort_values ( by = [("估值（亿人民币）","sum")], ascending = False) \
             .rename ( columns = {"sum":"总和", "mean":"均值", "count":"数量", "max":"最新", "min":"最早"} )
先国再城 = df.groupby ( by = ['国家','城市'] ) \
             .agg ({ "企业名称" : "count", \
                     "估值（亿人民币）":["sum","mean"], \
                     "成立年份":["max","min"],               }) \
             .sort_values ( by = [("估值（亿人民币）","sum")], ascending = False) \
             .rename ( columns = {"sum":"总和", "mean":"均值", "count":"数量", "max":"最新", "min":"最早"} )
先行再城 = df.groupby ( by = ['行业','城市'] ) \
             .agg ({ "企业名称" : "count", \
                     "估值（亿人民币）":["sum","mean"], \
                     "成立年份":["max","min"],               }) \
             .sort_values ( by = [("估值（亿人民币）","sum")], ascending = False) \
             .rename ( columns = {"sum":"总和", "mean":"均值", "count":"数量", "max":"最新", "min":"最早"} )
display(先国再行)
display(先行再国)
display(先国再城)
display(先行再城)

with pd.ExcelWriter("20春_pandas_week03_hurun_unicorn.xlsx") as writer:
    先国再行.to_excel(writer,sheet_name="先国再行") 
    先行再国.to_excel(writer,sheet_name="先行再国") 
    先国再城.to_excel(writer,sheet_name="先国再城") 
    先国再城.to_excel(writer,sheet_name="先行再城") 
```

```python
# 更改前
先行再城 = df.groupby ( by = ['行业','城市'] ) \
             .agg ({ "企业名称" : "count", \
                     "估值（亿人民币）":["sum","mean"], \
                     "成立年份":["max","min"],               }) \
             .sort_values ( by = [("估值（亿人民币）","sum")], ascending = False) \
             .rename ( columns = {"sum":"总和", "mean":"均值", "count":"数量", "max":"最新", "min":"最早"} )
display(先行再城)

# 更改后
先行再城 = df.groupby ( by = ['行业','城市'] ) \
             .agg ({ "成立年份":["max","min"],
                     "企业名称" : "count", \
                     "估值（亿人民币）":["sum","mean"], \
                                    }) \
             .sort_values ( by = [("估值（亿人民币）","sum")], ascending = False) \
             .rename ( columns = {"sum":"总和", "mean":"均值", "count":"数量", "max":"最新", "min":"最早"} )
display(先行再城)
```

```python
# 修改前
先国再城 = df.groupby ( by = ['国家','城市'] ) \
             .agg ({ "企业名称" : "count", \
                     "估值（亿人民币）":["sum","mean"], \
                     "成立年份":["max","min"],               }) \
             .rename ( columns = {"sum":"总和", "mean":"均值", "count":"数量", "max":"最新", "min":"最早"} )
display(先国再城)

# 修改后
先国再城 = df.groupby ( by = ['国家','城市'] ) \
             .agg ({ "企业名称" : "count", \
                     "估值（亿人民币）":["sum","mean"], \
                     "成立年份":["max","min"],               }) \
             .sort_values ( by = [("估值（亿人民币）","sum")], ascending = False) \
             .rename ( columns = {"sum":"总和", "mean":"均值", "count":"数量", "max":"最新", "min":"最早"} )
display(先国再城)

# .sort_values会将数值从高到低依次排列，这使得数据阅读更加轻松
```

#### 小结
* 报表3剑客:rename,sort_values,agg
* .rename能够更换观察observations对应的标签，可以让使用者换成自己更熟悉的文字，提高表格的可阅读性.
* .sort_values会将数值从高到低依次排列，这使得数据阅读更加轻松.
* .agg能够更换观察observations的排列顺序.

* 代码结尾的/表示这行代码已经结束了，代码后的/能够让机器更容易分析代码
